# Library Management System

Library Management System will consist of
1. Admin Panel – where only authorized admins will be logged in for managing
the store through id and password. Also, it includes the change password option.
2. Dashboard – it will include the name of the logged-in admin, date, and clock.
Also, it’s the main core of the project as it will be designed for 8 functionalities :
i. Add Books – add books to store
ii. Issue Books – issue books to students, max limit is 3, and the date of the last
issue book will be considered while returning
iii. Edit Books – edit details of the books
iv. Return Books – return the book and check for fine charges
v. Delete Books – delete books from the store
vi. Search Books – details about the book
vii. Show Books – show the whole book
viii. Log out – log out from the current admin session

```python
id, passwrd = 42240, 'studyaid'
books, issued = [], dict()
book_issued = []
```

```python
def inputs():
    global in_id, in_passwrd
    in_id = int(input('Enter the ID: '))
    in_passwrd = str(input('Enter the Password: '))
    return in_id, in_passwrd
inputs()
def dash(id, passwrd):
    def add_book():
        book_name = str(input('Enter The Name Of Book: '))
        books.append(book_name)
    def issue_book():
        stud_name = str(input('Enter The Name Of Student: '))
        issued_book = str(input('Enter The Book: '))
        if stud_name not in issued.keys():
            issued[stud_name] = [issued_book]
            book_issued.append(issued_book)
            books.remove(issued_book)
        else:
            if len(issued[stud_name]) <= 3:
                issued[stud_name].append(issued_book)
                book_issued.append(issued_book)
                books.remove(issued_book)
            else:
                print('You Cannot Issue More Books!')
    def edit_book():
        old_book = str(input('Enter The Book To Be Updated: '))
        books.remove(old_book)
        new_book = str(input('Enter The New Name Of Book: '))
        books.append(new_book)
    def return_book():
        ret_book = str(input('Enter The Name Of Book To Be Returned: '))
        books.append(ret_book)
        book_issued.remove(ret_book)
    def del_book():
        old_book = str(input('Enter The Book To Be Deleted: '))
        books.remove(old_book)
    def search_book():
        search = str(input('Enter The Book To Be Searched: '))
        if search in books:
            print(f'The Searched Book Is Available')
        else:
            print('The Book Not Found')
    def show_book():
        print(f'The Books Store Has Are: \n{books}')
    def log_out():
        id = int()
        passwrd = str()
    func = int(input('Enter 1 For Add_Book\n\
    Enter 2 For Issue_Book\n\
    Enter 3 For Edit_Book\n\
    Enter 4 For Return_Book\n\
    Enter 5 For Del_Book\n\
    Enter 6 For Search_Book\n\
    Enter 7 For Show_Book\n\
    Enter 8 For Log_Out: '))
    if func==1:
        add_book()
    elif func==2:
        issue_book()
    elif func==3:
        edit_book()
    elif func==4:
        return_book()
    elif func==5:
        del_book()
    elif func==6:
        search_book()
    elif func==7:
        show_book()
    elif func==8:
        log_out()
def admin_panel(id, passwrd):
    id = id
    passwrd = passwrd
    change_pass = int(input('Enter 1 To Change The Password or Enter 0: '))
    if change_pass == 1:
        passwrd = str(input('Enter the Password: '))
        dashboard = int(input('Enter 1 To View The Dashboard or Enter 0: '))
        if dashboard == 1:
            dash(id, passwrd)
        else:
            print('==============')
    else:
        dashboard = int(input('Enter 1 To View The Dashboard or Enter 0: '))
        if dashboard == 1:
            dash(id, passwrd)
        else:
            print('==============')
if in_id == id and in_passwrd == passwrd:
    admin_panel(id, passwrd)       
else:
    print('You Are Entering Invalid Infos')
    print('Kindly Try Again \n==================')
    inputs()
```
