# DESCRIPTION OF DATASET

'school': School the student attends (binary categorical: 'GP' or 'MS'). 'sex': Gender of the student (binary categorical: 'F' for female, 'M' for male). 'age': Age of the student (numeric). 'address': Type of address the student has ('U' for urban, 'R' for rural). 'famsize': Family size ('LE3' for less than or equal to 3, 'GT3' for greater than 3). 'Pstatus': Parent's cohabitation status ('T' for living together, 'A' for apart). 'Medu': Mother's education level (numeric). 'Fedu': Father's education level (numeric). 'Mjob': Mother's job. 'Fjob': Father's job. 'reason': Reason for choosing the school. 'guardian': Student's guardian. 'traveltime': Travel time to school (numeric). 'studytime': Weekly study time (numeric). 'failures': Number of past class failures (numeric). 'schoolsup': Extra educational support (binary categorical: 'yes' or 'no'). 'famsup': Family educational support (binary categorical: 'yes' or 'no'). 'paid': Extra paid classes within the course subject (binary categorical: 'yes' or 'no'). 'activities': Extra-curricular activities (binary categorical: 'yes' or 'no'). 'nursery': Attended nursery school (binary categorical: 'yes' or 'no'). 'higher': Wants to take higher education (binary categorical: 'yes' or 'no'). 'internet': Internet access at home (binary categorical: 'yes' or 'no'). 'romantic': In a romantic relationship (binary categorical: 'yes' or 'no'). 'famrel': Family relationship quality (numeric). 'freetime': Free time after school (numeric). 'goout': Going out with friends (numeric). 'Dalc': Workday alcohol consumption (numeric). 'Walc': Weekend alcohol consumption (numeric). 'health': Current health status (numeric). 'absences': Number of school absences (numeric). 'G1': First period grade (numeric). 'G2': Second period grade (numeric). 'G3': Final grade (numeric)

# RATIO BETWEEN SOME FEATURES

```python
fig,ax = pyp.subplots(5,2,figsize=(10,15))
ax[0][0].pie(data['sex'].value_counts(), labels=['Female', 'Male'], autopct='%1.1f%%', colors=['pink','lightgreen'],\
       shadow=True, explode=(0.06, 0))
ax[0][0].set_title('Distribution of Gender', weight='bold', fontsize=12)

ax[0][1].pie(data['address'].value_counts(), labels=['Urban', 'Rural'], autopct='%1.1f%%', colors=['goldenrod','gold'],\
       shadow=True, explode=(0.06, 0))
ax[0][1].set_title('Distribution of Addresses', weight='bold', fontsize=12)

ax[1][0].pie(data['guardian'].value_counts(), labels=['Mother','Father','Others'], autopct='%1.1f%%', colors=['goldenrod','gold','lightblue'],\
       shadow=True, explode=(0.06, 0.06, 0.03))
ax[1][0].set_title('Distribution of Guardian', weight='bold', fontsize=12)

ax[1][1].pie(data['studytime'].value_counts(), labels=['2Hr','3Hrs','1Hrs','4Hrs'], autopct='%1.1f%%', colors=['teal','lime','lavender','tomato'],\
       shadow=True, explode=(0.06, 0.03, 0.06, 0.06))
ax[1][1].set_title('Distribution of Guardian', weight='bold', fontsize=12)


failures_by_sex = data.groupby(['sex', 'failures']).size().unstack(fill_value=0)
ax[2][0].pie(failures_by_sex.loc['F'], labels=failures_by_sex.columns, autopct='%1.1f%%',\
           colors=['lime', 'orange', 'coral', 'cyan'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06))
ax[2][0].set_title('Female Failures Distribution', weight='bold', fontsize=12)

ax[2][1].pie(failures_by_sex.loc['M'], labels=failures_by_sex.columns, autopct='%1.1f%%',\
           colors=['lime', 'orange', 'coral', 'cyan'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06))
ax[2][1].set_title('Male Failures Distribution', weight='bold', fontsize=12)


medu_by_sex = data.groupby(['sex', 'Medu']).size().unstack(fill_value=0)
ax[3][0].pie(medu_by_sex.loc['F'], labels=medu_by_sex.columns, autopct='%1.1f%%',\
           colors=['grey', 'lightblue', 'lightgreen', 'brown'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06, 0.04))
ax[3][0].set_title('Female M_Edu Distribution', weight='bold', fontsize=12)

ax[3][1].pie(medu_by_sex.loc['M'], labels=medu_by_sex.columns, autopct='%1.1f%%',\
           colors=['grey', 'lightblue', 'lightgreen', 'brown'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06, 0.04))
ax[3][1].set_title('Male M_Edu Distribution', weight='bold', fontsize=12)


fedu_by_sex = data.groupby(['sex', 'Fedu']).size().unstack(fill_value=0)
ax[4][0].pie(fedu_by_sex.loc['F'], labels=fedu_by_sex.columns, autopct='%1.1f%%',\
           colors=['rosybrown', 'mediumpurple', 'lightblue', 'palegreen'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06, 0.04))
ax[4][0].set_title('Female F_Edu Distribution', weight='bold', fontsize=12)

ax[4][1].pie(fedu_by_sex.loc['M'], labels=fedu_by_sex.columns, autopct='%1.1f%%',\
           colors=['rosybrown', 'mediumpurple', 'lightblue', 'palegreen'], shadow=True, explode=(0.06, 0.03, 0.06, 0.06, 0.04))
ax[4][1].set_title('Male F_Edu Distribution', weight='bold', fontsize=12)
pyp.show()
```

# DESCRIPTIVE ANALYSIS RESPECT TO SEX

```python
fig,ax = pyp.subplots(4,1,figsize=(7.5,10))
ax[0].set_ylim(data['freetime'].min()-1, data['freetime'].max()+1)
sns.scatterplot(x='age',y='freetime',hue='sex',data=data,ax=ax[0], marker='o')
ax[0].set_ylabel('Freetime')

sns.countplot(x='age',hue='sex',data=data,ax=ax[1])
ax[1].set_ylabel('Age')

sns.countplot(x='age', data=data, hue='nursery', ax=ax[2], hue_order=data['nursery'].unique())
ax[2].set_ylabel('Nursery')

sns.boxplot(x='age', data=data, ax=ax[3])
ax[3].set_xlabel('Age')
pyp.show()
```
